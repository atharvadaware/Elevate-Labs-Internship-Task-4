```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, roc_auc_score
import warnings
warnings.filterwarnings('ignore')
```

## Separating features and target

```python
X = df.drop('label', axis= 1)
Y = df['label']
```

```python
print(X)
```

```python
print(Y)
```

## Splitting the data into training and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
```

# Model Training

## Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression model using training data

model.fit(X_train, Y_train)
```

# Model Evaluation

### Accuracy score

```python
# accuracy on training data

X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('Accuracy on training data : ', training_data_accuracy)
```

```python
# accuracy on testing data

X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('Accuracy on testing data : ', testing_data_accuracy)
```

### Confusion matrix


```python
# confusion matrix

conf_matrix = confusion_matrix(Y_test, X_test_prediction)
print('Confusion Matrix :\n', conf_matrix)
```

### Precision and Recall

```python
# calculate precision score
precision = precision_score(Y_test, X_test_prediction)
print('Precision on testing data : ', precision)

# calculate recall score
recall = recall_score(Y_test, X_test_prediction)
print('Recall on testing data : ', recall)
```

### ROC-AUC

```python
# calculate ROC-AUC score
roc_auc = roc_auc_score(Y_test, model.predict_proba(X_test)[:, 1])
print('ROC-AUC on testing data : ', roc_auc)
```

## Explain sigmoid function

## Tune threshold

```python
# Get predicted probabilities for the positive class
X_test_pred_prob = model.predict_proba(X_test)[:, 1]

# Create a range of thresholds
thresholds = np.arange(0, 1.01, 0.01)

# Store accuracy for each threshold
accuracies = []

# Iterate through each threshold and calculate accuracy
for threshold in thresholds:
    X_test_pred_threshold = (X_test_pred_prob >= threshold).astype(int)
    accuracy = accuracy_score(Y_test, X_test_pred_threshold)
    accuracies.append(accuracy)

# Display the thresholds and their corresponding accuracies
threshold_accuracy_df = pd.DataFrame({'Threshold': thresholds, 'Accuracy': accuracies})
display(threshold_accuracy_df.head())
display(threshold_accuracy_df.tail())
```

### Plot the accuracy against the thresholds to visualize the effect of threshold tuning.



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(threshold_accuracy_df['Threshold'], threshold_accuracy_df['Accuracy'])
plt.title('Accuracy vs. Prediction Threshold')
plt.xlabel('Threshold')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
```

```python
%%markdown
## The Sigmoid Function in Logistic Regression

The sigmoid function, also known as the logistic function, is a mathematical function that takes any real-valued number as input and maps it to a value between 0 and 1. The formula for the sigmoid function is:

$$ \sigma(x) = \frac{1}{1 + e^{-x}} $$

In the context of logistic regression, the sigmoid function is crucial because it transforms the output of the linear model into a probability. A linear model calculates a weighted sum of the input features and adds a bias term:

$$ z = w_1x_1 + w_2x_2 + ... + w_nx_n + b $$

where $w_i$ are the weights, $x_i$ are the input features, and $b$ is the bias.

The output $z$ can be any real number. However, for classification tasks, we need to output a probability that the input belongs to a certain class. This is where the sigmoid function comes in. By applying the sigmoid function to $z$, we get:

$$ P(Y=1|X) = \sigma(z) = \frac{1}{1 + e^{-z}} $$

This output, $P(Y=1|X)$, represents the probability that the instance $X$ belongs to the positive class (class 1), given the input features. The sigmoid function squashes the output of the linear model ($z$) into the range [0, 1], which is the valid range for a probability.

If $\sigma(z)$ is close to 1, it indicates a high probability of belonging to the positive class. If it's close to 0, it indicates a low probability of belonging to the positive class (and thus a high probability of belonging to the negative class). A common threshold, often 0.5, is then used to classify the instance into one of the two classes. If $\sigma(z) \ge 0.5$, the instance is classified as positive; otherwise, it's classified as negative.
```
